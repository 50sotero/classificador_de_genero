```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df1 = pd.read_csv('base_generos.csv')
```

```python
df1.columns = ['first_name', 'name', 'sex', 'frequency_female',
       'frequency_male', 'frequency_total', 'frequency_group', 'ratio',
       'alternative_names']
```

```python
df1.sex.unique()
```

```python
np.shape(df1)
```

```python
# Load our data
df2 = pd.read_csv('names_dataset.csv')
```

```python
df2.sex.unique()
```

```python
df = pd.concat([df1[['name','sex']], df2], ignore_index=True)
```

```python
df
```

```python
# Data Cleaning
# Checking for column name consistency
df.columns
```

```python
# Data Types
df.dtypes
```

```python
# Checking for Missing Values
df.isnull().isnull().sum()
```

```python
# Number of Female Names
df[df.sex == 'F'].size
```

```python
# Number of Male Names
df[df.sex == 'M'].size
```

```python
df_names = df
```

```python
# Replacing All F and M with 0 and 1 respectively
df_names.sex.replace({'F':0,'M':1},inplace=True)
```

```python
df_names.sex.unique()
```

```python
df_names.dtypes
```

```python
Xfeatures =df_names['name']
```

```python
# Feature Extraction 
cv = TfidfVectorizer()
X = cv.fit_transform(Xfeatures)
```

```python
cv.get_feature_names()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df_names.sex
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
vect = cv.transform(['Vinicius'])
```

```python
vect
```

```python
if clf.predict(vect)[0] == 1:
    print('M')
if clf.predict(vect)[0] == 0:
    print('F')
```
